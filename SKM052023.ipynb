## Вариант 2
Используя интеграл
###                                       <tex>$\pi=\int_0^1 \frac{4}{1+x^2}dx$</tex>
можно найти приближение к π.

**График подинтегральной функции:**

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
%matplotlib inline
x1 = np.arange(0, 1, 0.01)
y1 = 4 / (1 + x1**2)

plt.plot(x1, y1)
plt.grid()
```

### Выведем табулированные погрешности и значения интегральных сумм с графиками, используя правило прямоугольников, трапеций и Симпсона с  2 ,  4 ,  8 ,  16 ,  32 ,  64  и  128  узлами.

## Правило прямоугольника

```python
a = 0
b = 1
n = [2, 4, 8, 16, 32, 64, 128]
for uzl in n:
    h = (b - a)/uzl
    x = np.arange(a + h/2, b, h)
    y =4/(1+x**2)
    sum_y=sum(y)
    rectI =h*sum_y
    plt.plot(uzl,abs(np.pi-rectI)/(np.pi),'o',color='tab:red')
    print('При количестве узлов: ',uzl,' погрешность ',abs(np.pi-rectI)/(np.pi))
rectInaccuracy=abs(np.pi-rectI)/np.pi
plt.yscale('log')
plt.xscale('log')
pass
rectI
```

При удвоении числа элементарных отрезков погрешность для метода прямоугольника уменьшается в два раза.

## Правило трапеции 

```python
for uzl in n:
    h = (b - a)/uzl
    x = np.linspace(a, b, uzl + 1)
    y =4/(1+x**2)
    sum_y=sum(y)
    trapI =integrate.trapz(y, x)
    plt.plot(uzl,(np.pi-trapI)/(np.pi),'o',color='tab:red')
    print('При количестве узлов: ',uzl,' погрешность ',abs(np.pi-trapI)/(np.pi))
trapInaccuracy=abs(np.pi-trapI)/np.pi
plt.yscale('log')
plt.xscale('log')
pass
trapI

```

При удвоении числа элементарных отрезков погрешность для метода трапеции уменьшается в 4 раза.

## Правило Симпсона

```python
for uzl in n:
    h = (b - a)/uzl
    x = np.linspace(a, b, uzl + 1)
    y =4/(1+x**2)
    sum_y=sum(y)
    simpsonI =integrate.simps(y)*h
    plt.plot(uzl,abs(np.pi-simpsonI)/(np.pi),'o',color='tab:red')
    print('При количестве узлов: ',uzl,' погрешность ',abs(np.pi-simpsonI)/(np.pi))
simpsonInaccuracy=abs(np.pi-trapI)/np.pi
plt.yscale('log')
plt.xscale('log')
pass
simpsonI
```

При удвоении числа элементарных отрезков погрешность для метода Симпсона уменьшается в 16 раз.

В данных примерах погрешность функции постоянно уменьшается, но при дальнейшем увеличении узлов ошибка останется постоянной. Те вычисления становятся достаточно точными.

Сравним полученные результаты(интегральные суммы):

```python
print("Метод прямоугольника: ", rectI)
print("Метод трапеции: ", trapI)
print("Метод Симпсона: ", simpsonI)
print(np.pi)
```

**Вывод:** Самым эфективным оказался метод Сипсона, тк он получает наиболее приближенный результат за меньшее количество элементарных отрезков.
